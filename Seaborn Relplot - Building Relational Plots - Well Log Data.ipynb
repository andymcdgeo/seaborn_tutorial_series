# Seaborn Relational Plot (Relplot)

```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('Data/Xeek_Well_15-9-15.csv', na_values=-999)
```

```python
df.head()
```

## Scatter Plot with Relplot

```python
g = sns.relplot(data=df, x='NPHI', y='RHOB');
```

```python
g = sns.relplot(data=df, x='NPHI', y='RHOB')
g.set(ylim=(3, 1.5));
```

## Customising the Scatter Plot

```python
g = sns.relplot(data=df, x='NPHI', y='RHOB',
           hue='GR')
g.set(ylim=(3, 1.5));
```

```python
g = sns.relplot(data=df, x='NPHI', y='RHOB',
           hue='CALI')
g.set(ylim=(3, 1.5));
```

```python
g = sns.relplot(data=df, x='NPHI', y='RHOB',
           hue='LITH')
g.set(ylim=(3, 1.5));
```

```python
g = sns.relplot(data=df, x='NPHI', y='RHOB',
           hue='LITH',
           style='LITH')
g.set(ylim=(3, 1.5));
g.savefig('Lith Xplot.png', transparent=True, dpi=200)
```

## Line Plots

```python
g = sns.relplot(data=df, x='DEPTH_MD', y='GR', kind='line')
g.set(ylim=(0, 150));
```

```python
# Setting the size of the line plot

g = sns.relplot(data=df, x='DEPTH_MD', y='GR', kind='line', height=7, aspect=2)
g.set(ylim=(0, 150));
```

## Splitting Data into Facets

```python
g = sns.relplot(data=df, x='NPHI', y='RHOB',
           hue='LITH',
           style='LITH',
               col='LITH')
g.set(ylim=(3, 1.5));
```

```python
#Wrapping the columns

g = sns.relplot(data=df, x='NPHI', y='RHOB',
           hue='LITH',
           style='LITH',
               col='LITH', col_wrap=3)
g.set(ylim=(3, 1.5));
```

```python
#Changing size of points

g = sns.relplot(data=df, x='NPHI', y='RHOB',
           hue='LITH',
           style='LITH',
           col='LITH', col_wrap=3, 
           s=100)
g.set(ylim=(3, 1.5));
```
