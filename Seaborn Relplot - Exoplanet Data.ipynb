```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('Data/exoplanet_2021.csv')
```

```python
df.head()
```

```python
g = sns.relplot(data=df, x='Orbital Period Days', y ='Stellar Radius')
g.set(xlim=(0, 1000), ylim=(0, 100))
```

```python
g = sns.relplot(data=df, x='Orbital Period Days', y ='Stellar Radius', hue='Discovery Method')
g.set(xlim=(0, 10000), ylim=(0, 10))
```
